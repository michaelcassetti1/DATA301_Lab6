# Question 1

Find all cities with bike share programs in the United States (country code "US"), along with their network ID (for example, "ford-gobike"). How many cities in the U.S. have bike share programs (at least that are in this API)?

```python
import requests
import pandas as pd

json_data = requests.get("http://api.citybik.es/v2/networks").json()

cities = []
cities_id_pair = []

for network in json_data['networks']:
  if network['location']['country'] == "US":
    city = network['location']['city']
    network_id = network['id']
    cities.append(city)
    cities_id_pair.append((city, network_id))


print(len(cities_id_pair), "cities with bike share programs in the United States")
for city in cities_id_pair:
  print(city)
```

# Question 2

Construct a `DataFrame` containing data about all bike stations in all networks in the United States. Save this `DataFrame` to disk using `.to_csv()`. (You will need it in Part B of this lab.)

Use this `DataFrame` to determine the total number of bicycles in bike share programs across the United States? You may assume that the number of bikes at a station is the number of empty spaces, plus the number of available bikes.

```python
import time
us_stations = []

for city in cities_id_pair:
  url = "http://api.citybik.es/v2/networks/" + city[1]
  json_data = requests.get(url).json()
  #us_stations.append(json_data)
  #time.sleep(0.5)
  #print(json_data['network']['stations'])
  for station in json_data['network']['stations']:
    city = city
    name = station['name']
    free_bikes = station['free_bikes']
    empty_slots = station['empty_slots']
    timestamp = station['timestamp']
    longitude = station['longitude']
    latitude = station['latitude']
    id = station['id']
    station_list = {'city': city,
                    'name': name,
                    'free_bikes': free_bikes,
                    'empty_slots': empty_slots,
                    'timestamp': timestamp,
                    'longitude': longitude,
                    'latitude': latitude,
                    'id': id,}
    us_stations.append(station_list)


df_stations = pd.DataFrame(us_stations)
df_stations


```

```python
df_stations['total_bikes'] = df_stations['free_bikes'] + df_stations['empty_slots']
df_stations

print(df_stations['total_bikes'].sum(), "total bikes in the US")
```

```python
df_stations.to_csv('stations.csv', index = False)
```

# Question 3

You have just finished touring Coit Tower in San Francisco, which is located at latitude 37.802747 and longitude -122.405861. Using your `DataFrame` from Question 2, find the nearest bike station with an available bike, based on taxicab distance (a.k.a. Manhattan distance).

_Hint:_ You can check your answer using Google Maps!

```python
available_bike_stations = df_stations[df_stations['free_bikes'] > 0]
available_bike_stations['distance_from_Coit_Tower'] = (abs(37.802747 - available_bike_stations['latitude'])) + (abs((-122.405861 - available_bike_stations['longitude'])))
available_bike_stations = available_bike_stations.sort_values(by= 'distance_from_Coit_Tower')
available_bike_stations.iloc[0]
```
